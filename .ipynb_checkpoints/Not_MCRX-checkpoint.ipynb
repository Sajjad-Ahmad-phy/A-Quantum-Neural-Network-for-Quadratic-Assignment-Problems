```python
from qiskit import QuantumCircuit
from qiskit.circuit import Parameter
from qiskit.circuit.library.standard_gates import RXGate
import numpy as np

beta_para = Parameter(r'$\beta$')

# Define a custom gate using your parameter
class CustomRXGate(RXGate):
    def __init__(self, parameter):
        super().__init__(0)  # Initialize with a dummy parameter value
        self.parameter = parameter

    def to_matrix(self):
        # Check if the parameter is a ParameterExpression
        if isinstance(self.parameter, Parameter):
            cos = np.cos(self.parameter / 2)
            sin = np.sin(self.parameter / 2)
        else:
            cos = np.cos(float(self.parameter) / 2)
            sin = np.sin(float(self.parameter) / 2)
        return np.array([[cos, -1j * sin], [-1j * sin, cos]])

# Create a quantum circuit
qc = QuantumCircuit(5)

# Use your custom gate
MCRX = CustomRXGate(beta_para).control(num_ctrl_qubits=4, ctrl_state='0000')

# Append the gate to the circuit
qc.append(MCRX, [1, 2, 3, 4, 0])

qc.draw('mpl')
```
